# 1. Introduction to Natural Language Processing (NLP)

## Definition and importance of NLP

<div dir="rtl">

עיבוד שפה טבעית (באנגלית: Natural Language Processing - NLP)

הוא תת-תחום של בינה מלאכותית ובלשנות. הוא עוסק בבעיות הקשורות לעיבוד ומניפולציה של שפה טבעית והבנה של שפה טבעית על מנת לגרום למחשבים "להבין" דברים שנאמרים או נכתבים בשפות אנושיות.

</div>
